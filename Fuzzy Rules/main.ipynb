## Fuzzy Rules

### Libs

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Triangle MF

```python
def triangular(x, a, b, c):
    if x<=a:
        return 0
    elif a<=x<=b:
        return (x-a)/(b-a)
    elif b<=x<=c:
        return (c-x)/(c-b)
    else:
        return 0 
```

```python
z = np.linspace(0, 10, 1000)

C1 = np.array([triangular(val, 2, 4, 6) for val in z])
C2 = np.array([triangular(val, 5, 7, 9) for val in z])

alpha1 = 0.6
alpha2 = 0.3
```

```python
C1_prime = np.minimum(C1, alpha1)
C2_prime = np.minimum(C2, alpha2)

C_final = np.maximum(C1_prime, C2_prime)
```

```python
plt.plot(z, C1,  linestyle="--")
plt.plot(z, C2, linestyle="--")
plt.plot(z, C1_prime,  color='blue')
plt.plot(z, C2_prime,  color='green')
plt.plot(z, C_final,  color='red',)
plt.ylabel("Membership Degree")
```
